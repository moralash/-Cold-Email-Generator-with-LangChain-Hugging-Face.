```python
from langchain.llms import HuggingFaceHub
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
email_template = """
You are an AI that writes professional cold emails.

Write a cold email to {recipient_name}, a {role} at {company}.
Introduce our product, {product_name}, and explain how it helps with {product_benefit}.
Our goal is to {goal}.
Keep the tone friendly and persuasive.
"""

prompt = PromptTemplate(
    input_variables=["recipient_name", "company", "role", "product_name", "product_benefit", "goal"],
    template=email_template
)
```

```python
from langchain.llms import HuggingFaceHub

llm = HuggingFaceHub(
    huggingfacehub_api_token=hf_token,  # Pass token directly here
    repo_id="mistralai/Mistral-7B-Instruct-v0.1",  # or another supported model
    model_kwargs={"temperature": 0.7, "max_new_tokens": 300}
)
```

```python
email_chain = LLMChain(llm=llm, prompt=prompt)
```

```python
email_input = {
    "recipient_name": "Emily Green",
    "company": "TechNova",
    "role": "CTO",
    "product_name": "SmartHire AI",
    "product_benefit": "streamlining the hiring process with AI-based screening",
    "goal": "schedule a quick product demo this week"
}
```

```python
cold_email = email_chain.run(email_input)
print(cold_email)
```
